```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.DataFrame([[180000,110,18.9,1400],[360000,905,23.4,1800],[230000,230,23.4,1800],[60000,450,13.5,1500],[700000,630,14.5,2000]],columns=['colA', 'Col B', 'Col C', 'Col D' ])
```

```python
print(df)
```

```python
df.plot(kind ='bar')
```

```python
max_scaled = df.copy() # if we have to copy our data into another variable then we have to use copy() command
```

```python
print(max_scaled)
```

```python
for column in max_scaled.columns:
  max_scaled[column] = max_scaled[column] / max_scaled[column].abs().max()

  print(max_scaled)

  max_scaled.plot(kind = 'bar')
  plt.show()
```

MIN MAX FEATURE SCALING

```python
scaled = df.copy()

for column in scaled.columns:
    scaled[column] = (scaled[column] - scaled[column].min()) / (scaled[column].max() - scaled[column].min())

print(scaled)
scaled.plot(kind='bar')
plt.show()
```

```python
z_scaled = df.copy()

for column in z_scaled.columns:
    z_scaled[column] = (z_scaled[column] - z_scaled[column].mean()) / z_scaled[column].std()

print(z_scaled)

z_scaled.plot(kind='bar')
plt.show()
```
